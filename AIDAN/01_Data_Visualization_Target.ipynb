# Visualization on the Zillow Rent Index

```python
df_metro_fl=zri[zri["Metro"].isin(["Miami-Fort Lauderdale-West Palm Beach",
                       "Orlando-Kissimmee-Sanford", 
                       "Tampa-St. Petersburg-Clearwater",
                      "Jacksonville"])]
```

```python
df_metro_fl.columns.tolist()
#There're 7 columns which are not "months"
```

```python
df_metro_fl.shape #145 zip codes in all 4 metro areas 
```

```python
df_metro_fl
```

```python
df_metro_fl_trans=df_metro_fl.T
```

```python
df_metro_fl_trans
```

```python
df_metro_fl_trans.columns = df_metro_fl_trans.iloc[1]
```

```python
df_metro_fl_trans
```

```python
zri_by_fl_zip=df_metro_fl_trans.drop(["RegionID", 
                                "RegionName", 
                                "City", "State", "Metro", 
                                "CountyName", "SizeRank"], axis=0)
```

```python
zri_by_fl_zip
```

```python
zri_by_fl_zip.columns = zri_by_fl_zip.columns.astype(str)
```

```python
plt.figure(figsize = (32, 16))
plt.plot(zri_by_fl_zip.index, zri_by_fl_zip["33160"])
plt.xticks(np.arange(4,112, 12), range(2011, 2020, 1), size=45)
plt.yticks(size=40)
plt.title("Zillow Rent Index ($) Over Time for the Zip Code 33160", size=50)
plt.show()
```

### Making seperate dataframe for each of the 4 metro areas

# Max, Min and Average Zillow Rent Index Over Time: Miami

```python
df_miami=df_metro_fl[df_metro_fl["Metro"]=="Miami-Fort Lauderdale-West Palm Beach"]
```

```python
df_miami.head()
```

```python
df_miami.groupby("Metro").mean()
```

```python
#Taking only the "months" columns
df_miami_mean=df_miami.groupby("Metro").mean().iloc[: , 3:]
df_miami_mean=df_miami_mean.T
```

```python
df_miami_mean.columns=["Mean_Rent"]
df_miami_mean
```

```python
df_miami_max=df_miami.groupby("Metro").max().iloc[: , 6:].T
```

```python
df_miami_max.columns=["Max_Rent"]
```

```python
df_miami_max
```

```python
df_miami_min=df_miami.groupby("Metro").min().iloc[: , 6:].T
df_miami_min.columns=["Min_Rent"]
df_miami_min
```

```python
df_miami_mean.shape
```

```python
plt.figure(figsize = (32, 16))

plt.plot(df_miami_max.index, df_miami_max["Max_Rent"], color='r')
plt.plot(df_miami_mean.index, df_miami_mean["Mean_Rent"], color='b')
plt.plot(df_miami_min.index, df_miami_min["Min_Rent"], color='g')

plt.xticks(np.arange(4,112, 12), range(2011, 2020, 1), size=45)
#speificy where the first instance starts, 5th month in this case 
#plt.xticks(rotation=45, size=45)
plt.yticks(size=40)
plt.title("Max, Min and Average Zillow Rent Index ($) Over Time for Miami", size=50)
plt.legend(['Max','Average','Min'], prop={'size': 40})
sns.set_style("darkgrid")

plt.show()
```
