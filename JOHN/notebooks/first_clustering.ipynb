```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
total_df = pd.read_csv('../../JACOB/total_df')
total_df
```

```python
# for s in total_df.columns:
#     print(s)
```

```python
acs_2018 = total_df[ total_df['year'] == 2018 ].copy()
acs_2018
```

```python
kmeans_cols = [ 'zip_code', 'gini_index', 'median_age' ]
kmeans_df = acs_2018[ kmeans_cols ]
kmeans_input = kmeans_df.drop('zip_code', axis=1)
```

```python
km = KMeans( n_clusters=5, random_state=0, tol=0.01 ).fit(kmeans_input)
# print(type(km.labels_))
kmeans_df['cluster_label'] = km.labels_
kmeans_df
```
