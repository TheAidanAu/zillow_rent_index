```python
bin_strings = ['income_less_10000','income_10000_14999', 'income_15000_19999',
                 'income_20000_24999','income_25000_29999','income_30000_34999',
                 'income_35000_39999','income_40000_44999','income_45000_49999',
                 'income_50000_59999','income_60000_74999','income_75000_99999',
                 'income_100000_124999','income_125000_149999','income_150000_199999',
                 'income_200000_or_more']

#over200K midpoint is estimated to be 300k,
#shouldn't matter for median calc
bin_midpoints = [5000,12500,17500,22500,27500,32500,
                 37500,42500,47500,55000,67500,87500,
                 112500,137500,175000,300000]

len(bin_strings), len(bin_midpoints)

household_by_bins, total_households = [], 0

for s in bin_strings:
    total_households += tetris[s].sum()
    household_by_bins.append(tetris[s].sum())

# household_by_bins, total_households

cum_sum = 0
for idx in range(len(household_by_bins)):
    cum_sum += household_by_bins[idx]
    if cum_sum >= total_households / 2:
        break

bin_midpoints[idx]
    
```
