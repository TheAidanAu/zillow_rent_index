```python
summary_df_2015 = df_2015[[
            'white_pop', 'total_pop', 'gini_index', 'commuters_by_bus', 'children',
            'median_rent', 'median_income', 'in_undergrad_college',
            'vacant_housing_units_for_rent', 'vacant_housing_units_for_sale', 'nonfamily_households'
    ]]
```

```python
summary_df_2015 = summary_df_2015.dropna()
print(summary_df_2015.shape)

summary_df_2015_Y = summary_df_2015[['median_rent']]
summary_df_2015_X = summary_df_2015.drop('median_rent', axis=1 )
summary_df_2015_X.shape, summary_df_2015_Y.shape
```

```python
from sklearn import linear_model
ols = linear_model.LinearRegression()
ols.fit(summary_df_2015_X, summary_df_2015_Y)
```

```python
print('Intercept: \n', ols.intercept_)
print('Coefficients: \n', ols.coef_)
```
