```python
import pandas as pd
import numpy as np
```

```python
multi_index = pd.read_csv('multifam.csv')
```

```python
multi_index.shape
```

```python
index_fl = multi_index.loc[multi_index['State']== 'FL']
```

```python
columns = index_fl.columns
date_columns = []
other_columns = []
for column in columns:
    if '2' in column:
        date_columns.append(column)
    else:
        other_columns.append(column)
print(date_columns)
print(other_columns)
index_fl.shape
```

```python
fl = pd.melt(index_fl, id_vars = other_columns , value_vars = date_columns)
fl = fl.rename(columns={"variable": "Date", "value": "Rent_Index"})
```

```python
fl
```

```python
fl['RegionName'].unique(), len(fl['RegionName'].unique())
```

```python
fl['RegionID'].unique(), len(fl['RegionID'].unique())
```

```python
fl['Date'] = pd.to_datetime(fl['Date'])
```

```python
fl['Year'] = fl.Date.dt.year
fl['Yearly_Avg'] = fl.groupby([fl.Metro, fl.Year])['Rent_Index'].transform('mean')
```

```python
print(fl.Metro.unique())
Metros_Used = ['Miami-Fort Lauderdale-West Palm Beach','Orlando-Kissimmee-Sanford', 'Tampa-St. Petersburg-Clearwater','Tallahassee']
fl = fl.loc[fl['Metro'].isin(Metros_Used)]
```
