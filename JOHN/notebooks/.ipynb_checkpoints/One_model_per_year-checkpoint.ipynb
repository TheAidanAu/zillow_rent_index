```python
wide_df = pd.pivot_table( total_df, index='zip_code', columns='year')
```

```python
wide_df.shape
```

```python
for s in wide_df.columns:
    print(s)
```

```python
summary_df = wide_df[[
            ('white_pop', 2017), ('vacant_housing_units_for_rent', 2018), ('gini_index', 2016),
            ('median_rent', 2016), ('median_rent', 2017), ('median_rent', 2018),
            ('median_income', 2016), ('median_income', 2017), ('median_income', 2018),
    ]]
```

```python
summary_df.shape, summary_df.dtypes
```

```python
summary_df = summary_df.dropna()
print(summary_df.shape)

summary_df_Y = summary_df[[('median_rent', 2018)]]
summary_df_X = summary_df.drop( ('median_rent', 2018), axis=1 )
summary_df_X.shape, summary_df_Y.shape
```

```python
from sklearn import linear_model
ols = linear_model.LinearRegression()
ols.fit(summary_df_X, summary_df_Y)
```

```python
print('Intercept: \n', ols.intercept_)
print('Coefficients: \n', ols.coef_)
```
