```python
def no_seq_nans(df):
    output = []
    for i1 in range(df.shape[0]):
        currently_nan = False
        for i2 in range(7, 120):
            if df.iloc[i1, i2]:
                currently_nan = False
            elif currently_nan:
                output.append(False)
                break
            else:
                currently_nan = True
        else:
            output.append(True)
    return output
```
